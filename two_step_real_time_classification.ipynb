# Real-time classification

```python
import cv2
import numpy as np
import tensorflow as tf
import winsound
```

```python
model = tf.keras.models.load_model('cnn_trained_model1')
```

```python
class_names = ['fall detected','walk','sit']
```

```python
cap = cv2.VideoCapture(0)
```

```python
alert_frequency = 1000
alert_duration = 1000
```

```python
while True:
    ret, frame = cap.read()

    # text_labels = []
    frame = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
    frame = cv2.resize(frame, (160, 160))
    frame = frame / 127.5 - 1.0

    frame = np.expand_dims(frame, axis = 0)

    predictions = model.predict(frame)
    predicted_class_index = np.argmax(predictions)
    predicted_class = class_names[predicted_class_index]

    cv2.putText(frame, predicted_class, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (O, 255, 0), 2)

    if predicted_class_index == 2;
      winsound.Beep(alert_frequency, alert_duration)
    
    cv2.imshow('Real Time Classfication', frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
