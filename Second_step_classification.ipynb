```python
import os
import cv2
import numpy as np
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import joblib

# 数据文件夹路径
data_folder = '/content/gdrive/MyDrive/dataset'

# 遍历文件夹并加载图像数据
X = []
y = []
for category in os.listdir(data_folder):
    category_folder = os.path.join(data_folder, category)
    if os.path.isdir(category_folder):
        for image_name in os.listdir(category_folder):
            image_path = os.path.join(category_folder, image_name)
            image = cv2.imread(image_path)
            image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  # 将图像转为灰度图
            image = cv2.resize(image, (64, 64))  # 调整图像大小
            feature_vector = image.flatten()  # 将图像展平为特征向量
            X.append(feature_vector)
            y.append(category)

# 将数据转为NumPy数组
X = np.array(X)
y = np.array(y)

# 数据拆分为训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# 特征标准化
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# 创建SVM分类器
clf = svm.SVC(kernel='linear')

# 训练分类器
clf.fit(X_train, y_train)

# 保存模型
joblib.dump(clf, 'svm_model.pkl')

# 在测试集上进行预测
predictions = clf.predict(X_test)

# 评估模型性能
accuracy = clf.score(X_test, y_test)
print("Accuracy:", accuracy)
```

```python
import cv2
import numpy as np
from sklearn.preprocessing import StandardScaler
import joblib
from google.colab.patches import cv2_imshow

# 加载保存的模型
model_path = '/content/svm_model.pkl'
clf = joblib.load(model_path)

# 准备输入数据（照片集）
photo_folder = '/content/gdrive/MyDrive/dataset/fall'

# 遍历照片文件夹并处理每张照片
for photo_name in os.listdir(photo_folder):
    photo_path = os.path.join(photo_folder, photo_name)
    image = cv2.imread(photo_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  # 转为灰度图
    image = cv2.resize(image, (64, 64))  # 调整图像大小
    feature_vector = image.flatten()  # 将图像展平为特征向量

    # 特征标准化
    scaler = StandardScaler()
    feature_vector_scaled = scaler.fit_transform(feature_vector.reshape(1, -1))

    # 使用模型进行预测
    prediction = clf.predict(feature_vector_scaled)

    # 显示图片和预测结果
    cv2_imshow(image)
    print("Photo:", photo_name)
    print("Prediction:", prediction)

```
